```python
cars_numeric=df.select_dtypes(include=['int64','float64'])
cars_numeric
```

```python
cars_numeric=cars_numeric.drop(columns=['car_ID','symboling'],axis=1) # axis=1 means entire column is dropped
cars_numeric
```

```python
x=cars_numeric.iloc[:,0:13]

y=cars_numeric.iloc[:,13].values
# y=cars_numeric['price'].values
```

```python
#TRAIN TEST SPLIT ON THE DATASET
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
# x_train,y_train used to fitting the model
# x_test is used for predicting
# y_test is used for comparing actual product with y_pred
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```
