```python
x=np.arange(0,50,0.1)
y=np.sin(x)
plt.plot(x,y,color="blue")
plt.fill_between(x,y,color="red",alpha=0.5)
plt.show()
```

```python
x=np.linspace(-1,1,20)
y=np.exp(x)
plt.scatter(x,y,c=y)
plt.colorbar()
```

```python
#LINE GRAPHS
a=[1,2,3,4,5,6]
b=[5,4,6,8,3,2]
c=[1,4,7,9,6,2]

plt.plot(a,b,color="red",label="line 1")
plt.plot(a,c,color="blue",label="line 2")
plt.title("LINE GRAOH")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(title="lines",loc="upper right")
plt.grid()
plt.show()
```

```python
a=[1,2,3,4,5,6]
b=[5,4,6,8,3,2]
c=[1,4,7,9,6,2]

plt.plot(a,b,color="red",label="line 1",linestyle="dashed",marker="^",markersize="10",markerfacecolor="green")
plt.plot(a,c,color="blue",label="line 2",linestyle="dotted",linewidth=3)
plt.title("LINE GRAOH")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(title="lines",loc="upper right")
plt.grid()
plt.show()
```

```python
#BAR GRAPH
name=['AA','BB','CC','DD','EE']
weights=[40,30,20,60,10]
color1=['r','g','b','c','m']
plt.bar(name,weights,color=color1)
```

```python
name=['AA','BB','CC','DD','EE']
weights=[40,30,20,60,10]
color1=['r','g','b','c','m']
plt.title("HORIZOTAL BAR GRAPH",fontsize=15)
plt.barh(name,weights,color=color1)
```

```python
#PIE CHART
#percentage of gases in the air
gases=['nitrogen','oxygen','CO2','other gases']
percentage=[50,40,2,8]
color1=['red','green','blue','yellow']
plt.pie(percentage,labels=gases,colors=color1,explode=[0,0.1,0,0.2],startangle=90,counterclock=False,autopct='%1.2f%%',shadow=True)
plt.show()
plt.savefig("pie.png")
```

```python
#SCATTER PLOT
a=[9,8,7,6,5,4,3,2,1]
b=[1,4,7,9,8,2,6,3,5]
ax=plt.axes()
ax.set_facecolor('r')
plt.scatter(a,b,color="b",s=200,marker=">")
```
